# Диаграммы рассеяния - Scatter Plots
Диаграммы рассеяния (график рассеяния, график разброса) показывают, как различные переменные зависят друг от друга - это основная идея всех графиков, которые показывают связь между переменными.  

Графики Scatter Plots показывают связь между двумя непрерывными переменными. **"Непрерывная"** переменная - это числовая переменная, которая может принимать любые промежуточные значения. Такие переменные позволяют точнее детализировать то или иное значение. *Примеры*: возраст, высота, зарплата, температура.

В непрерывной переменной между двумя значениями всегда найдётся какое-то третье значение. Это сильно отличается от категориальных переменных, где каждое значение - отдельная категория. *Примеры*: цвета, имена, формы и т.д.

Такие связи можно показать с помощью разных типов графиков. На примере `scatterplot()` посмотрим на различные параметры seaborn, применимые и для других графиков.

График Scatter Plots берёт две непрерывные переменные, и наносит их на две оси графика (прим.: зарплата и объем продаж).

---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Данные

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
df.info()
```

---

## Диаграмма рассеяния - Scatterplot

```python
sns.scatterplot(data=df, x='salary', y='sales')
plt.show()
```

## Подключение к графику с помощью Matplotlib
**Важно: matplotlib самостоятельно подключается к графику seaborn (даже без импортирования matplotlib.pyplot). Причина в том, что seaborn сам по себе создает объект Figure в matplotlib. При желании можно импортировать matplotlib.pyplot и вызвать команды matplotlib, обращаясь к графику seaborn.**

```python
plt.figure(figsize=(12, 4), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales')
plt.show()
```

## Параметры Seaborn
Очень часто в seaborn используются следующие параметры - hue и palette.

### Параметр hue
Раскрашивает точки на графике, используя значения некоторой категориальной переменной в объекте DataFrame. **В качестве данного параметра обычно указывается категориальная переменная.**

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', hue='division')
plt.show()
```

Если указать непрерывную переменную в качестве аргумента параметра hue, то seaborn автоматически создаст градиент цвета с различными градациями цвета. В легенде будут отображаться некоторые пороговые значения.

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', hue='work experience')
plt.show()
```

### Параметр palette
Значения параметра palette можно выбрать из карты цветов (cmap) в Matplotlib.  
https://matplotlib.org/tutorials/colors/colormaps.html

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', hue='work experience', palette='viridis')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', hue='level of education', palette='Dark2')
plt.show()
```

## Параметры Scatterplot
Указанные ниже параметры в большей мере относятся именно к scatterplot()

### Параметр size
Позволяет задать размер точек на основе значений некоторой колонки. Для данного параметра можно выбрать как непрерывную, так и категориальную переменную.

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', size='salary')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', size='work experience')
plt.show()
```

### Параметр s
С помощью s= можно задать одинаковый размер маркера.

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', s=200)
plt.show()
```

### Параметр alpha
Прозрачность точек

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', s=200, alpha=0.3)
plt.show()
```

### Парметр linewidth
Контур точек

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', s=200, linewidth=0)
plt.show()
```

### Парметр style
Для настройки маркеров. Автоматически выбирает стили на основе некоторой атегориальной переменной в наборе данных. Также можно с помощью параметра **markers=** передать список типов маркеров на основе matplotlib, например: ['*','+','o']

```python
plt.figure(figsize=(12, 8), dpi=100)
sns.scatterplot(data=df, x='salary', y='sales', style='level of education')
plt.show()
```

```python
plt.figure(figsize=(12, 8), dpi=100)
# Иногда очень удобно задать ОБА параметра - hue и style - на основе одной и той же колонки
sns.scatterplot(data=df, x='salary', y='sales', style='level of education', hue='level of education', s=100)
plt.show()
```

## Экспортируем график Seaborn

```python
plt.figure(figsize=(12, 8), dpi=100)
# Иногда очень удобно задать ОБА параметра - hue и style - на основе одной и той же колонки
sns.scatterplot(data=df, x='salary', y='sales', style='level of education', hue='level of education', s=100)

# Вызываем savefig в той же ячейке
plt.savefig('example_scatter.jpg')

plt.show()
```
