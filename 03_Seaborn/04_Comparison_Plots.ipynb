# Графики сравнения
---
Такие графики являются двухмерными версиями рассмотренных ранее графиков. 

Выделим два типа графиков:
* jointplot()
* pairplot()



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

## График jointplot()
С помощью данного метода мы можем взять гистограммы для каждой переменной графика scatterplot(), чтобы уточнить распределение данных по каждой переменной. Кроме scatterplot() можно использовать hexplot() (шестиугольники на таком графике темнее там, где больше точек) или двумерный kdeplot().

```python
# Выберем две непрерывные переменные
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score')
plt.show()
```

### Параметр kind для графиков jointplot()

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', kind='hex')
plt.show()
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', kind='scatter', alpha=0.2)
plt.show()
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', kind='hist')
plt.show()
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', kind='kde')
plt.show()
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', kind='kde', fill=True)
plt.show()
```

```python
plt.figure(dpi=200)
sns.jointplot(data=df, x='math score', y='reading score', hue='gender', palette='viridis')
plt.show()
```

## График pairplot()
Это очень быстрый способ сравнивать между собой все числовые переменные в наборе данных. Автоматически создает гистограмму для каждой переменной, и сравнительные графики scatterplot() для всех комбинаций переменных. 

**Важно: такой график может потреблять много CPU и памяти в случае больших объемов данных и большого количества колонок. Лучше - сначала оставить только те колонки, которые нужно исследовать.**

```python
# Не потребовалось выбирать отдельные колонки, так как
# в наборе числовых колонок всего три
plt.figure(dpi=200)
sns.pairplot(data=df)
plt.show()
```

```python
# Графики KDE по умолчанию на главной диаганали при указании параметра hue
plt.figure(dpi=200)
sns.pairplot(data=df, hue='gender', palette='viridis')
plt.show()
```

```python
plt.figure(dpi=200)
sns.pairplot(data=df, hue='gender', palette='viridis', diag_kind='hist')
plt.show()
```

```python
# Убираем дублирующиеся графики
plt.figure(dpi=200)
sns.pairplot(data=df, hue='gender', palette='cividis', corner=True)
plt.show()
```
