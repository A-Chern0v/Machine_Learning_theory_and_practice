# Сетки - Grids
Seaborn grid - это способ построения набора графиков рядом друг с другом. В основе лежит `subplots()` из Matplotlib, с помощью которого автоматически создаётся набор графиков для категориальной переменной.  
Указывать количество колонок и строк не нужно, вместо этого необходимо указать название колонки, и seaborn автоматически создаёт набор графиков.

---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

## Метод .catplot()

```python
plt.figure(dpi=200)
sns.catplot(data=df, x='gender', y='math score', hue='gender', kind='box')
plt.show()
```

```python
# В данном примере hue - помещает графики на одну пару осей
plt.figure(dpi=200)
sns.catplot(data=df, x='gender', y='math score', hue='lunch', kind='box')
plt.show()
```

```python
# Параметр row - создаст отдельные пары осей 
# для каждой из категорий
plt.figure(dpi=200)
sns.catplot(data=df, x='gender', y='math score', kind='box', hue='gender', row='lunch')
plt.show()
```

```python
plt.figure(dpi=200)
sns.catplot(data=df, x='gender', y='math score', kind='box', hue='gender', col='lunch')
plt.show()
```

```python
plt.figure(dpi=200)
sns.catplot(data=df, x='gender', y='math score', kind='box', hue='gender', row='test preparation course', col='lunch')
plt.show()
```

```python
plt.figure(dpi=200)
sns.catplot(data=df, x='gender', y='math score', kind='violin', hue='gender', row='test preparation course', col='lunch')
plt.show()
```

## Метод .pairgrid()

```python
plt.figure(dpi=200)
sns.PairGrid(data=df)
plt.show()
```

```python
plt.figure(dpi=200)

g = sns.PairGrid(data=df)
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot, color='aqua')
g = g.map_diag(sns.histplot)

plt.show()
```

```python
plt.figure(dpi=200)

g = sns.PairGrid(data=df, hue='gender', palette='viridis')
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot, color='aqua')
g = g.map_diag(sns.histplot)
g = g.add_legend()

plt.show()
```

## Метод FacetGrid

```python
sns.FacetGrid(data=df, col='gender', row='lunch')
plt.show()
```

```python
g = sns.FacetGrid(data=df,col='gender',row='lunch')
g = g.map(plt.scatter, "math score", "reading score", edgecolor="w")
g.add_legend()
plt.show()
```

```python
g = sns.FacetGrid(data=df,col='gender',row='lunch')
g = g.map(plt.scatter, "math score", "reading score", edgecolor="w")
g.add_legend()

plt.subplots_adjust(hspace=0.4, wspace=1)
```
