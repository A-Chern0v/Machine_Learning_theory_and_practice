После того, как были рассмотрены базовые графики, можно сделать более формальный подход к объектно-ориентированному API для Matplotlib. Будем создавать объекты **figure**, и затем обращаться к атрибутам и методам этих объектов.

## Создаем объект Figure

Основная идея - создавать объекты и обращаться к их атрибутам и методам. Такой подход более удобен в случаях, когда на одном рисунке должны находиться несколько графиков.

В основе лежит объект Figure, который представляет собой некоторое пустое пространство на экране, на которое можно добавлять оси или несколько пар осей и рисовать один или несколько графиков. Это позволяет очень детально контролировать весь процесс создания и настройки графиков.  
Чтобы увидеть объект Figure на экране, необходимо добавить в него хотя бы одну пару осей. По умолчанию размер объекта 640x480px.

1. Создаем объект Figure, сохраняем результат в переменную
2. Для объекта Figure добавляем пару осей и сохраняем результат в переменную
3. Для осей вызываем метод plot() чтобы нарисовать график

```python
# Создаем пустое поле для рисования (canvas/figure)
fig = plt.figure()

# Добавляем набор осей (при таких параметрах полностью повторится ширина и высота объекта Figure)
axes = fig.add_axes([0, 0, 1, 1]) # left, bottom, width, height (диапазон от 0 до 1)

# Наносим график на этот набор осей
axes.plot(x, y)

plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(a, b)
plt.show()
```

## Добавляем ещё один набор осей
Для одного объекта Figure можно добавить несколько наборов осей с разными размерами и распложением. Для каждого из наборов осей можно рисовать отдельные графики.

```python
type(fig)
```

В данном пример гравик a, b будет нарисован два раза

```python
# Создаем поле для рисования
fig = plt.figure()

axes_1 = fig.add_axes([0, 0, 1, 1])    # Большой набор осей
axes_2 = fig.add_axes([0.2, 0.2, 0.5, 0.5])    # Маленький набор осей

# Рисуем график на бОльшем наборе осей - axes_1
axes_1.plot(a, b)

# С помощью методов set_ указываем атрибуты для осей
axes_1.set_xlabel('X Label')
axes_1.set_ylabel('Y Label')
axes_1.set_title('Big Figure')

# Рисуем график на меньшем наборе осей - axes_2
axes_2.plot(a, b)
axes_2.set_title('Small Figure')

plt.show()
```

### Можно подвинуть меньший набор осей и поменять его параметры

```python
fig = plt.figure()

# Набор больших осей
axes_1 = fig.add_axes([0, 0, 1, 1])
axes_1.plot(a, b)
axes_1.set_xlim(0, 8)
axes_1.set_ylim(0, 8000)
axes_1.set_xlabel('A')
axes_1.set_ylabel('B')
axes_1.set_title('Возведение в сетепень 4')

# Набор маленькиз осей
axes_2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes_2.plot(a, b)
axes_2.set_xlim(1, 2)
axes_2.set_ylim(0, 50)
axes_2.set_xlabel('A')
axes_2.set_ylabel('B')
axes_2.set_title('Увеличенный фрагмент')

plt.show()
```

### Можно добавить несколько наборов осей, даже за пределами основного набора осей 

```python
fig = plt.figure()

axes_1 = fig.add_axes([0, 0, 1, 1])              # Большой набор осей
axes_2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])    # Маленький набор осей
axes_3 = fig.add_axes([1, 1, 0.25, 0.25])        # Набор осей в правом верхнем углу!

# Рисуем график на бОльшем наборе осей - axes_1
axes_1.plot(a, b)

# С помощью методов set_ указываем атрибуты для осей
axes_1.set_xlabel('X Label')
axes_1.set_ylabel('Y Label')
axes_1.set_title('Big Figure')

# Рисуем график на меньшем наборе осей - axes_2
axes_2.plot(a, b)
axes_2.set_xlim(8, 10)
axes_2.set_ylim(4000, 10000)
axes_2.set_xlabel('X')
axes_2.set_ylabel('Y')
axes_2.set_title('Zoomed In')

# Рисуем график на третьем наборе осей - axes_3
axes_3.plot(a, b)

plt.show()
```

## Параметры объекта Figure

    dpi - увеличение разрешения для объекта Figure
    figsize - размер объекта Figure (измеряется в дюймах)

```python
fig = plt.figure(figsize=(12, 8), dpi=100)

axes_1 = fig.add_axes([0, 0, 1, 1])

axes_1.plot(a, b)

plt.show()
```

## Экспорт объекта Figure
    bbox_inches='tight' автоматически подбирает правильный прямоугольник вокруг графика

```python
fig = plt.figure(figsize=(12, 8), dpi=100)

axes_1 = fig.add_axes([0, 0, 1, 1])

axes_1.plot(a, b)
axes_1.set_label('X')

fig.savefig('figure.png', bbox_inches='tight')

plt.show()
```
