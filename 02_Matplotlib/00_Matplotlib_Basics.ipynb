## Основы Matplotlib

Matplotlib является "основополагающей" библиотекой по визуализации данных в Python. Эта библиотека позволяет создавать двумерные и трёхмерный графики.

Основные преимущества Matplotlib следующие:
* Достаточно просто начать создавать простые графики
* Возможность создавать свои собственные текстовые надписи на графиках
* Возможность настраивать любые элементы на графике
* Экспорт результатов в разных форматах, с высоким качеством
* Множество других настроек графиков

Matplotlib позволяет программно создавать графики, которые можно при необходимости воспроизвести заново.  
Оффициальная документация библиотеки: http://matplotlib.org/

## Функция plt.plot()
### Базовые команды Matplotlib

Самый простой способ рисовать графики в Matplotlib (функциональный способ) - с помощью функции plot:  
`plt.plot(x,y)`  
Такие вызовы функций просты, однако не позволяют детально настраивать параметры графиков. Их рекомендуется использовать для быстрой визуализации данных.

```python
plt.plot(x, y)
```

### Редактирование внешнего вида графика

```python
plt.plot(x, y)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Название графика')
plt.show() # Также позволяет убрать префикс Out[]
```

```python
plt.plot(x, y)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.xlim(0, 6) # Нижняя граница, верхняя граница
plt.ylim(0, 12) # Нижняя граница, верхняя граница
plt.title('Название графика')
plt.show()
```
