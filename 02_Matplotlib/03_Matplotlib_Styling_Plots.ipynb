## Легенда
С помощью именованного параметра **label=label_text** можно указать названия для тех или иных линий на графике. Далее с помощью метода **legend**, уже без араметров, можно добавить легенду.

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x, label='X vs X')

ax.plot(x, x**2, label='X vs X^2')

ax.legend()

plt.show()
```

Потенциально, легенда может загородить собой часть линий. 

В методе **legend** есть опциональный именованный параметр **loc**, с помощью которого можно указать, где именно рисовать легенду. Возможные значения параметра **loc** - это числа, каждое из которых соответствует определённому месту на графике. Более подробно можно почитать [здесь](http://matplotlib.org/users/legend_guide.html#legend-location). Наиболее частые значения **loc** следующие:

```python
# Много разных вариантов ...

ax.legend(loc=1) # Верхний правый угол
ax.legend(loc=2) # Верхний левый угол
ax.legend(loc=3) # Нижний левый угол
ax.legend(loc=4) # Нижний правый угол

# ... доступны также другие варианты

# Хороший вариант
ax.legend(loc=0) # matplotlib сам выбирает оптимальное местоположение
fig
```

```python
ax.legend(loc=(1.1, 0.5)) # задаем вручную
fig
```

## Цвета, ширина линий и типы линий
Matplotlib предоставляет *много* возможностей по настройке цветов, ширины и типов линий.

Для этого можно использовать базовый синтаксис, похожий на MATLAB (именно MATLAB, а не Matplotlib). Но этот синтаксис не очень удобен, и лучше избегать его. 

### Быстрый обзор:

### Цвета с использованием синтаксиса MatLab
В matplotlib можно задавать цвета линий и других графических элементов разными способами. Первый способ - похожий на MATLAB синтаксис, в котором `'b'` означет blue, `'g'` означает green и так далее. Также поддерживается MATLAB API для выбора стиля линий:
например, `'b.-'` означает цвет blue и линия с точками:

```python
# Цвета и стили с применением синтаксиса MATLAB
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
plt.show()
```

## Рекомендуемый способ: именованные параметры
### Для цветов есть параметр color
Второй способ - указывать цвета с помощью их названий, а также с помщью шестнадцатиричных кодов RGB. Оптимально можно ещё указать прозрачность. Параметры называются `color` и `alpha`.

```python
fig, ax = plt.subplots()

ax.plot(x, x+1, color='blue', alpha=0.5)    # полупрозрачный
ax.plot(x, x+2, color='#8B008B')            # RGB hex code
ax.plot(x, x+3, color='#FF8C00')            # RGB hex code

plt.show()
```

## Стилизация линий и маркеров
### Ширина линий
Для изменения ширины линии есть именованный параметр `linewidth` или `lw`.

```python
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

# Можно указывать как linewidth, так и lw
ax.plot(x, x-1, color='red', linewidth=0.25)
ax.plot(x, x-2, color='red', lw=0.50)
ax.plot(x, x-3, color="red", lw=1)
ax.plot(x, x-4, color='red', linewidth=10)

plt.show()
```

### Стили линий
Есть много разных вариантов для `linestyle` или `ls`, некоторые из них:

```python
# Возможные варианты для параметра ls: '--', '-', '-.', ':'

fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, x-1, color='green', lw=3, linestyle='-')    # solid
ax.plot(x, x-2, color='green', lw=3, ls='-.')          # dash and dot
ax.plot(x, x-3, color='green', lw=3, ls=':')           # dots
ax.plot(x, x-4, color='green', lw=3, ls='--')          # dashes

plt.show()
```

### Произвольная стилизация линии
При желании можно задать любую последоватеьность отрезков и пробелов между ними. Для этого можно указать длину каждого отрезка (в количестве точек) и длину пробела между каждым из отрезков (также в количестве точек).

Нпример, (5, 2, 1, 2) будет обозначать следующее:

    отрезок линии длиной в 5 точек
    затем пробел длиной в 2 точки
    далее отрезок линии длиной в 1 точку
    и наконец, пробел длиной в 2 точки.

Для начала посмотрим, как можно получить линию с помощью команды `.plot()`:

```python
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

lines = ax.plot(x, x)

plt.show()

print(type(lines))
```

```python
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)
# Произвольная стилизация линии
lines = ax.plot(x, x+8, color='black', lw=5)
lines[0].set_dashes([10, 10])    # формат: длина линии, длина пробела.

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)
# Произвольная стилизация линии
lines = ax.plot(x, x+8, color='black', lw=5)
lines[0].set_dashes([1,1, 1,1, 10,10])    # формат: длина линии, длина пробела и т.д.

plt.show()
```

### Маркеры
По сути мы всегда рисовали точки, и matplotlib автоматически соединял эти точки с помощью линий. Как можно нарисовать маркеры в каждой из этих точек.

### Стили маркеров
Большой список различных типов маркеров можно найти [здесь](https://matplotlib.org/3.2.2/api/markers_api.html)

```python
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

# Параметр marker содержит строковый код
# Параметр markersize или ms задаёт размер

ax.plot(x, x-1, marker='+', markersize=20)
ax.plot(x, x-2, marker='o', ms=20)          # ms = markersize
ax.plot(x, x-3, marker='s', ms=20, lw=0)    # указываем linewidth=0, чтобы увидеть только маркеры
ax.plot(x, x-4, marker='1', ms=20)

plt.show()
```

### Различные края маркеров, ширина, размер и стиль

```python
fig, ax = plt.subplots(figsize=(12, 6), dpi=100)

# Рахмер и цвет маркеров
ax.plot(x, x, color='black', lw=1, ls='-', marker='s', markersize=20,
       markerfacecolor='green', markeredgewidth=8, markeredgecolor='blue')

plt.show()
```
