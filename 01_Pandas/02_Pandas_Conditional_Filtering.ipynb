# Фильтрация данных на основе условий (Conditional Filtering)

В больших массивах удобнее фильтровать данные не по позиции элеиентов, а по некоторому условию. "Фильтрация по условию" позволяет
выбрать __строки__, накладывая условия на значения __колонок__.  
__Колонки__ - это признаки, а признаки - это атрибуты, которые являются некоторым свойством или характеристикой каждой строки.  
__Строки__ - это отдельные экземпляры данных, для которых есть набор признаков.  
__Такой формат необходим для машинного обучения!__  

Варианты фильтрации по условию:
- Фильтр по одному условию
- Фильтр по нескольким условиям
- Проверка нескольких условий

1. Для начала необходимо выбрать колонку в датафрейме, например __df["Pop"]__.
2. Затем, применяем условие, на основе которого происходит фильтрация __df["Pop"] > 50__. Pandas выполнить операцию сравнения в режиме
broadcast, применяя сравнения для каждого отдельного элемента в колонке. В результате такого сравнения Pandas вернет объект __Series__
со значением True или False для каждого экземпляра данных.
3. Теперь, на основе этих значений можно выбрать из датафрейма только те строки, где условие вернуло True: __df[df["Pop"]>50]__.
Таким образом мы получаем объект DataFrame.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
df.head()
```

*__Выбор по условию для одной колонки__*

```python
# Для начал выберем строку.
df['total_bill']
```

```python
# Проведем сравнение и получим объект Series.
df['total_bill'] > 40
```

```python
# Теперь можно сделать так:
bool_series = df['total_bill'] > 40
df[bool_series]
```

```python
# Либо можно сделать в один шаг:
df[df['total_bill'] > 40]
```

```python
df[df['sex'] == 'Male']
```

```python
df[df['time'] != 'Dinner']
```

*__Выбор на основе нескольких условий__*

AND -> & - оба условия будут выполнены -> (условие_1) & (условие_2)  
OR -> | - хотя бы одно условие выполняется -> (условие_1) | (условие_2)

AND и OR написать нельзя, так как они применяются к отдельным элементам, поэтому используется синтаксис Pandas чтобы операция 
выполнилась в режиме broadcast

```python
df[(df['total_bill'] > 30) | (df['sex'] == 'Male')]
```

```python
df[(df['smoker'] == 'Yes') & (df['sex'] == 'Female')]
```

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat') | (df['day'] == 'Fri')]
```

Для трех условий список становится слишком громоздким. Вместо этого можно сделать следующим образом:

```python
# Список с условиями для выбора.
days = ['Fri', 'Sat', 'Sun']
# Получаем объект Series.
df['day'].isin(days)
```

```python
# Выполяем фильтрацию датафрейма на основе условия
df[df['day'].isin(days)]
```

```python
# Можно также выполнить всё в одной строке
df[df['day'].isin(['Sat', 'Sun'])]
```
